### Traversing List

````python
def traversinglst(lst):
    for idx, val in enumerate(lst):
        print(lst[idx], end = ' ')
````

````python
aquatic_animals = ['üêã', 'üêüÔ∏è', 'ü¶ë', 'üê°', 'ü¶à']

traversinglst(aquatic_animals)
````

### Updating List

````python
aquatic_animals[2] = 'üêô'

traversinglst(aquatic_animals)
````

### List Method

````python
aquatic_animals.insert(2, 'ü¶ë')
traversinglst(aquatic_animals)
````

````python
aquatic_animals.append('üêã')
traversinglst(aquatic_animals)
````

````python
pet_animals = ['üê∂', 'üê∞']

aquatic_animals.extend(pet_animals)
animals = aquatic_animals 
traversinglst(animals)
````

### Slicing & Deletion

````python
pet_animals = aquatic_animals[7:]
traversinglst(pet_animals)
````

````python
pet_animals.pop() # default value of index = -1
traversinglst(pet_animals)
````

````python
del aquatic_animals[7:]
traversinglst(aquatic_animals)
````

````python
aquatic_animals.remove('üêô')
traversinglst(aquatic_animals)
````

### Searching in List

````python
print('üêüÔ∏è' in aquatic_animals, 'The fish exists!!')
````

````python
def searchInList(lst, value):
    try:
        for val in lst:
            if val == value:
                return lst.index(val)
    except:
        print('The value didn\'t exists in list')
````

````python
searchInList(aquatic_animals, 'üêüÔ∏è')
````
